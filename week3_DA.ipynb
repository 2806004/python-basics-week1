```python
import pandas as pd
df=pd.read_csv('/content/BMW sales data (2010-2024) (1).csv')
pd.info(df)
```

```python
df.head()
```

```python
df['TOTAL_SALES']=df['Price_USD']*df['Sales_volume']
df.head()
```

```python
df['Total_Revenue']=df['TOTAL_SALES'].max()
df.head()
```

```python
df.drop("Best_selling_product",axis=1,inplace=True)
df.head()
```

```python
best_selling_product = df.groupby("Model")["Sales_Volume"].sum().sort_values(ascending=False).index[0]
print(best_selling_product)
```

```python
print(df)
Total_Revenue = df['TOTAL_SALES'].max()
print("Total Revenue:", Total_Revenue)
print("Best Selling Product:",best_selling_product )
```
